### Importing the required libraries

```python
from pandas import read_csv
import numpy as np
```

### Importing the dataset

```python
dataset = read_csv(r'pima-indians-diabetes.csv', header=None)
```

```python
print("------ Printing 10 headers of the dataset ------")
dataset.head(10)
```

#### Clearly, missing values (0) can be seen in the given dataset.

```python
print("\n------ Summarizing the dataset ------")
dataset.info()
```

```python
print("\n------ Counting the number of missing values in each column ------")
(dataset[list(range(0,9))] == 0).sum()
```

#### As can be seen, columns 0 through 5 all have missing values in them. Thus, we will perform missing value replacement only in the aforementioned columns.

```python
print("\n------ Replacing 0 values with NaN ------")
for column in range(6):
    dataset[column].replace(0, np.nan, inplace=True)
dataset.head(10)
```

#### The missing zero values have now been replaced with 'NaN' in the above dataset.

#### Now, we will replace the NaN values with the mean of the respective column.

```python
print("\n------ Replacing NaN values with Mean of respective column ------")
for column in range(6):
    mean_value = dataset[column].mean()
    dataset[column] = dataset[column].fillna(mean_value)
dataset.head(10)
```

#### The missing values have hence been removed and replaced.
